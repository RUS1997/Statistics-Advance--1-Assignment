#Q1. What is a random variable in probability theory?
  - A random variable is a set if possible values from random expriment.
#Q2. What are the types of random variables?
  - Discrete: Takes countable values with positive probability.
  - Continuous: Takes values in intervals (uncountable) with density.
#Q3. What is the difference between discrete and continuous distributions?
  - Discrete: Defined by a PMF, gives P(X=x), positive for specific values
  - Continuous: Defined by a PDF, where exact values have zero probability; probability is area under curve over intervals
#Q4. What are probability distribution functions (PDF)?
  - For continuous variables, a probability density function f(x) describes density such that [P(a<X<b)=∫f(x)dx].It is non-negative and integrates to 1.
  - Draw Outcome in a form of distribution.
#Q5. How do CDFs differ from PDFs?
  - A cumulative distribution function (CDF) F(x)=P(X≤x) applies to all random variables.
#Q6. What is a discrete uniform distribution?
  - A discrete uniform distribution assigns equal probability to a finite set of values a,a+1,...,b; each has probability 1/(b−a+1).
#Q7. Key properties of a Bernoulli distribution.
  - It's discrete with support {0,1}.
  - PMF: P(X=1)=p, P(X=0)=1−p.
  - Mean = p, variance = p(1−p).
#Q8. What is the binomial distribution, and how is it used?
  - It describes the number of successes in n independent Bernoulli trials with success probability p. PMF:P(X=k)=p^k(1−p)^(n−k).

#Q10. What is the Poisson distribution and applications?
  - A discrete distribution modeling counts of events in fixed intervals when events occur independently at constant rate λ. PMF: P(X=k)=e^−λ(λ^K/k!). Used in modeling rare events like phone calls, accidents, or incoming traffic.
#Q11. Characteristics of a normal distribution
  - Symmetric, bell-shaped, unimodal.
  - Defined by mean μ and variance σ^2.
  - PDF: proportional to e^−(x−μ)^2/(2σ^2).
  - Support covers entire real line.
#Q12. What is the standard normal distribution, and why is it important?
  - A special normal distribution with mean 0 and variance 1. It’s crucial because other normal variables can be standardized (Z-scores) and tabulated using this distribution.
#Q13. What is the Central Limit Theorem (CLT), and why is it critical?
  - The CLT states that, for sufficiently large sample size n, the sum (or average) of independent identically distributed random variables tends toward normal distribution, regardless of the original distribution. It underpins many statistical methods and inference techniques.
#Q14. What is the application of Z statistics in hypothesis testing?
  - You compute a Z-statistic to test hypotheses about population means (when σ known or sample large), comparing Z= (X −μ)/(σ/√n) to standard normal critical values or compute p-values.
#Q15. How to calculate a Z‑score, and what does it represent?
  - A Z-score measures how many standard deviations x is from the mean—quantifies the position of a data point within a distribution.
  Z= (X-μ)/σ.
#Q16. What are point estimates and interval estimates?
  - Point estimate: a single best guess of a parameter (e.g., sample mean X for population mean).
  - Interval estimate: a range (confidence interval) around the point estimate that likely contains the parameter.
#Q17. Significance of confidence intervals
  - A confidence interval (e.g., 95% CI) quantifies uncertainty in estimates; it defines a range which, over repeated sampling, contains the true parameter in the stated percentage of cases, offering meaningful inferential insight.
#Q18. Relationship between Z-score and confidence intervaL.
  - Confidence intervals are constructed using Z-scores: e.g., for mean with known zα/2*(σ/√n).
#Q19. How are Z‑scores used to compare different distributions?
  - Z‑scores standardize values from different distributions by converting them to the same scale (mean 0, std 1), enabling meaningful comparison across diverse distributions.
#Q20. Assumptions for applying the CLT.
  - Observations must be independent.
  - Distribution should have finite mean and variance.
  - Larger n improves approximation, especially if the original distribution is non-normal.
#Q21. Concept of expected value
  - The expected value E[X] is the long-run average of a random variable:
  - Discrete: ∑xP(X=x)
  - Continuous: ∫xf(x)dx
#Q22. How distribution relates to expected outcome
  - A probability distribution (PMF/PDF) weights each possible outcome based on probability. The expected value is the probability-weighted average, representing the variable’s long-run average behavior.

```python
#Q2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

low, high = 1, 7
rv = st.randint(low, high)
values = np.arange(low, high)
pmf = rv.pmf(values)

plt.bar(values, pmf, align='center', alpha=0.6)
plt.xlabel('k'); plt.ylabel('P(X=k)')
plt.title('PMF of Discrete Uniform Distribution')
plt.show()
```

```python
#Q3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pmf(k, p):
    return p**k * (1 - p)**(1 - k)  # k = 0 or 1

# Example usage:
print(bernoulli_pmf(1, 0.3), bernoulli_pmf(0, 0.3))
```

```python
#Q4. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram.
import numpy as np
import matplotlib.pyplot as plt

n, p = 10, 0.5
data = np.random.binomial(n, p, size=10000)

plt.hist(data, bins=np.arange(n+2)-0.5, density=True, rwidth=0.8)
plt.xlabel('Number of successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.show()
```

```python
#Q5. Create a Poisson distribution and visualize it using Python.
from numpy.random import poisson
import matplotlib.pyplot as plt
import seaborn as sns

lam = 4
samples = poisson(lam, size=10000)
sns.histplot(samples, stat='density', discrete=True, bins=np.arange(0, max(samples)+1)-0.5)
x = np.arange(0, np.max(samples)+1)
from scipy.stats import poisson as p_dist
plt.plot(x, p_dist.pmf(x, lam), 'bo-', label='PMF')
plt.title('Poisson Distribution (λ=4)')
plt.xlabel('k'); plt.ylabel('P(X=k)')
plt.legend(); plt.show()
```

```python
#Q6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

low, high = 1, 7
rv = st.randint(low, high)
x = np.arange(low, high)
cdf = rv.cdf(x)

plt.step(x, cdf, where='post')
plt.xlabel('k'); plt.ylabel('CDF')
plt.title('CDF of Discrete Uniform Distribution')
plt.show()
```

```python
#Q9. Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt

def z_scores(arr):
    return (arr - np.mean(arr)) / np.std(arr)

data = np.random.normal(50, 5, size=100)
zs = z_scores(data)

plt.stem(zs)
plt.xlabel('Index'); plt.ylabel('Z‑score')
plt.title('Z‑scores of Dataset')
plt.show()
```

```python
#Q10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt

pop = np.random.exponential(scale=1.0, size=100000)
sample_size = 30
num_samples = 1000
means = [np.mean(np.random.choice(pop, sample_size)) for _ in range(num_samples)]

plt.hist(means, bins=30, density=True, alpha=0.7)
plt.title('Sampling Distribution of the Mean\n(CL T Demo with Exponential Population)')
plt.xlabel('Sample mean'); plt.ylabel('Density')
plt.show()
```

```python
#Q20. Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
from scipy import stats

data = np.random.normal(50, 8, size=40)
m, s, n = data.mean(), data.std(ddof=1), len(data)
ci = stats.t.interval(0.95, df=n-1, loc=m, scale=s/np.sqrt(n))
print("95% CI:", ci)
```

```python
#Q21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
import numpy as np
from scipy import stats

np.random.seed(42)
mu_true, sigma_true = 100, 15
n = 50
data = np.random.normal(mu_true, sigma_true, size=n)

x_bar = data.mean()
s = data.std(ddof=1)

alpha = 0.05
df = n - 1
t_crit = stats.t.ppf(1 - alpha/2, df)

moe = t_crit * (s / np.sqrt(n))
ci_lower, ci_upper = x_bar - moe, x_bar + moe

print(f"Sample mean = {x_bar:.2f}")
print(f"95% CI = ({ci_lower:.2f}, {ci_upper:.2f})")
```
